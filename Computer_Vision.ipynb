   #**Experiment - 1(Filtering)**

```python
#!pip install opencv-python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/flow.jpg')

if img is None:
    print("Error: Image not found or unable to load.")
else:
    id_kernel = np.array([[0, 0, 0],
                          [0, 1, 0],
                          [0, 0, 0]])
    flt_img = cv2.filter2D(src=img, ddepth=-1, kernel=id_kernel)
    flt_img_rgb = cv2.cvtColor(flt_img, cv2.COLOR_BGR2RGB)

    plt.imshow(flt_img)
    plt.title('Identity Filter Result')
    plt.axis('off')  # Hide the axes
    plt.show()
```

#**Experiment - 2(Canny)**

```python
#!pip install opencv-python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/flow.jpg')

if img is None:
    print("Error: Image not found or unable to load.")
else:
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred_img = cv2.GaussianBlur(gray_img, (5, 5), 0)
    edges_canny = cv2.Canny(blurred_img, 100, 200)

    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Canny edge detection
    plt.subplot(1, 2, 2)
    plt.imshow(edges_canny,cmap='gray')
    plt.title('Canny Edge Detection')
    plt.axis('off')

    plt.show()
```

#**Experiment - 3(Sobel)**

```python
#!pip install opencv-python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

img = cv2.imread('/content/flow.jpg')

if img is None:
    print("Error: Image not found or unable to load.")
else:
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=5)  # X direction
    sobel_y = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=5)  # Y direction
    sobel_combined = cv2.magnitude(sobel_x, sobel_y)

    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    # Sobel edge detection
    plt.subplot(1, 2, 2)
    plt.imshow(sobel_combined, cmap='gray')
    plt.title('Sobel Edge Detection')
    plt.axis('off')

    plt.show()
```

#**Experiment - 5(Harris)**

```python
#!pip install opencv-python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/test.jpg')
if image is None:
    print("Error: Image not found or unable to load.")
else:
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  gray_float = np.float32(gray_image)

  harris_response = cv2.cornerHarris(gray_float, 2, 5, 0.07)
  harris_image = image.copy()
  harris_image[harris_response > 0.01 * harris_response.max()] = [0, 0, 255]

  cv2_imshow(harris_image)  # Harris Corner Detection
```

#**Experiment - 9(Haar_Cascades)**

```python
import cv2
from google.colab.patches import cv2_imshow

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

img = cv2.imread('/content/in.jpg')

if img is None:
    print("Error: Image not found or unable to load.")
else:
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Detect objects (faces in this case)
    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    # Draw rectangles around the detected objects
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

    cv2_imshow(img)
```

#**Experiment - 10(KCF) - !!In Local**

```python
#pip install opencv-python==4.5.5.64 opencv-contrib-python==4.5.5.64
import cv2
tracker = cv2.legacy.TrackerKCF_create()
video = cv2.VideoCapture('in.mp4')
ret, frame = video.read()

if not ret:
    print("Failed to read the video file. Exiting...")
    video.release()
    cv2.destroyAllWindows()
    exit()

bbox = cv2.selectROI("Select Object to Track", frame, False)
tracker.init(frame, bbox)

while True:
    ret, frame = video.read()
    if not ret:
        break

    success, bbox = tracker.update(frame)
    if success:
        x, y, w, h = map(int, bbox)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
    else:
        cv2.putText(frame, "Tracking failure", (100, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0, 0, 255), 2)

    cv2.imshow("Object Tracking", frame)
    # Exit on ESC key
    if cv2.waitKey(1) & 0xFF == 27:
        break
video.release()
cv2.destroyAllWindows()
```

#**Experiment - 14**

```python
from time import time
import matplotlib.pyplot as plt
from scipy.stats import loguniform
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Load dataset
lfw_people = fetch_lfw_people(min_faces_per_person=30, resize=0.4)
X, y = lfw_people.data, lfw_people.target
target_names = lfw_people.target_names

print(f"Total dataset size:\nn_samples: {X.shape[0]}\nn_features: {X.shape[1]}\nn_classes: {target_names.shape[0]}")

# Train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
# Scale the data
scaler = StandardScaler()
X_train, X_test = scaler.fit_transform(X_train), scaler.transform(X_test)

# PCA
n_components = min(150, X_train.shape[0], X_train.shape[1])
pca = PCA(n_components=n_components, svd_solver="randomized", whiten=True).fit(X_train)
X_train_pca, X_test_pca = pca.transform(X_train), pca.transform(X_test)

# Train classifier
param_grid = {"C": loguniform(1e3, 1e5), "gamma": loguniform(1e-4, 1e-1)}
clf = RandomizedSearchCV(SVC(kernel="rbf", class_weight="balanced"), param_grid, n_iter=10)
clf.fit(X_train_pca, y_train)

# Predictions
y_pred = clf.predict(X_test_pca)
print(classification_report(y_test, y_pred, target_names=target_names))
ConfusionMatrixDisplay.from_estimator(clf, X_test_pca, y_test, display_labels=target_names, xticks_rotation="vertical")
plt.tight_layout()
plt.show()

# Plotting functions
def plot_gallery(images, titles, h, w, n_row=3, n_col=4):
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())

# Titles for predictions with added spacing
prediction_titles = [f"predicted: {target_names[y_pred[i]].rsplit(' ', 1)[-1]}\n\ntrue: {target_names[y_test[i]].rsplit(' ', 1)[-1]}" for i in range(y_pred.shape[0])]
plot_gallery(X_test, prediction_titles, lfw_people.images.shape[1], lfw_people.images.shape[2])
# Plot eigenfaces
eigenfaces = pca.components_.reshape((n_components, lfw_people.images.shape[1], lfw_people.images.shape[2]))
plot_gallery(eigenfaces, [f"eigenface {i}" for i in range(eigenfaces.shape[0])], lfw_people.images.shape[1], lfw_people.images.shape[2])
plt.show()
```

#**Experiment - 14.1**

```python
from time import time
import matplotlib.pyplot as plt
from scipy.stats import loguniform
from sklearn.datasets import fetch_lfw_people
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# Load dataset
lfw_people = fetch_lfw_people(min_faces_per_person=30, resize=0.4)
X, y = lfw_people.data, lfw_people.target
target_names = lfw_people.target_names

print(f"Total dataset size:\nn_samples: {X.shape[0]}\nn_features: {X.shape[1]}\nn_classes: {target_names.shape[0]}")

# Train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
# Scale the data
scaler = StandardScaler()
X_train, X_test = scaler.fit_transform(X_train), scaler.transform(X_test)

# Apply Fisherfaces using LDA
n_components = min(target_names.shape[0] - 1, X_train.shape[1])  # At most (n_classes - 1) components
lda = LDA(n_components=n_components)
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)

# Train classifier
param_grid = {"C": loguniform(1e3, 1e5), "gamma": loguniform(1e-4, 1e-1)}
clf = RandomizedSearchCV(SVC(kernel="rbf", class_weight="balanced"), param_grid, n_iter=10)
clf.fit(X_train_lda, y_train)

# Predictions
y_pred = clf.predict(X_test_lda)
print(classification_report(y_test, y_pred, target_names=target_names))
ConfusionMatrixDisplay.from_estimator(clf, X_test_lda, y_test, display_labels=target_names, xticks_rotation="vertical")
plt.tight_layout()
plt.show()

# Plotting functions
def plot_gallery(images, titles, h, w, n_row=3, n_col=4):
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())

# Titles for predictions with added spacing
prediction_titles = [f"predicted: {target_names[y_pred[i]].rsplit(' ', 1)[-1]}\n\ntrue: {target_names[y_test[i]].rsplit(' ', 1)[-1]}" for i in range(y_pred.shape[0])]
plot_gallery(X_test, prediction_titles, lfw_people.images.shape[1], lfw_people.images.shape[2])

# Plot Fisherfaces
fisherfaces = lda.scalings_.reshape((n_components, lfw_people.images.shape[1], lfw_people.images.shape[2]))
plot_gallery(fisherfaces, [f"fisherface {i}" for i in range(fisherfaces.shape[0])], lfw_people.images.shape[1], lfw_people.images.shape[2])
plt.show()
```

#**Experiment - 15**


```python
import cv2
import numpy as np

# Load the video or webcam feed
cap = cv2.VideoCapture(r'D:\Kishore--Workspace\CV&IP\videoplayback1.mp4')  # Replace with 0 to use the webcam

# Lucas-Kanade parameters
lk_params = dict(winSize=(15, 15), maxLevel=2,
                 criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

# Read the first frame
ret, old_frame = cap.read()
if not ret:
    print("Failed to read video")
    exit()

old_gray = cv2.cvtColor(old_frame, cv2.COLOR_BGR2GRAY)

# Shi-Tomasi corner detection parameters
p0 = cv2.goodFeaturesToTrack(old_gray, mask=None, maxCorners=100, qualityLevel=0.3, minDistance=7, blockSize=7)

# Create a mask image for drawing optical flow
mask = np.zeros_like(old_frame)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Calculate Optical Flow using Lucas-Kanade method
    p1, st, err = cv2.calcOpticalFlowPyrLK(old_gray, frame_gray, p0, None, **lk_params)

    # Select good points
    good_new = p1[st == 1]
    good_old = p0[st == 1]

    # Draw the tracks
    for i, (new, old) in enumerate(zip(good_new, good_old)):
        a, b = new.ravel()  # Flattening the array
        c, d = old.ravel()

        # Convert to integer for drawing
        a, b = int(a), int(b)
        c, d = int(c), int(d)

        # Draw line and circle
        mask = cv2.line(mask, (a, b), (c, d), color=(0, 255, 0), thickness=2)
        frame = cv2.circle(frame, (a, b), 5, color=(0, 0, 255), thickness=-1)

    # Overlay the original frame with the mask showing the flow
    img = cv2.add(frame, mask)

    # Show the frame
    cv2.imshow('Optical Flow - Lucas-Kanade', img)

    # Exit if ESC is pressed
    if cv2.waitKey(30) & 0xFF == 27:
        break

    # Update previous frame and points
    old_gray = frame_gray.copy()
    p0 = good_new.reshape(-1, 1, 2)

cap.release()
cv2.destroyAllWindows()
```
